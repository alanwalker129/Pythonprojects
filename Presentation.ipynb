```python
import fitz  # PyMuPDF
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import time

def display_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    page_count = len(doc)

    for page_num in range(page_count):
        page = doc.load_page(page_num)
        pix = page.get_pixmap()
        img = np.array(pix.samples).reshape((pix.height, pix.width, 3))

        plt.imshow(img)
        plt.title(f"Page {page_num + 1} of {page_count}")
        plt.axis('off')
        plt.draw()
        plt.pause(1)  # Short pause for the transition effect
        plt.clf()  # Clear the current figure
        time.sleep(120)  # Wait for 2 minutes before showing the next page

    plt.close()

if __name__ == "__main__":
    pdf_path = "/content/Ai-Presentation.pdf"
    display_pdf(pdf_path)
```

```python
import fitz  # PyMuPDF
import matplotlib.pyplot as plt
import numpy as np
import time

def display_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    page_count = len(doc)

    plt.ion()  # Enable interactive mode
    fig, ax = plt.subplots()

    start_time = time.time()  # Record the start time

    for page_num in range(page_count):
        page = doc.load_page(page_num)
        pix = page.get_pixmap()

        # Check if pix.samples is valid
        if pix.samples is None or len(pix.samples) == 0:
            print(f"Warning: Page {page_num + 1} has no pixel data.")
            continue

        # Ensure correct shape and number of channels
        img = np.frombuffer(pix.samples, dtype=np.uint8).reshape((pix.height, pix.width, pix.n))

        ax.imshow(img)
        ax.set_title(f"Page {page_num + 1} of {page_count}")
        ax.axis('off')
        plt.draw()
        plt.pause(1)  # Short pause for the transition effect

        # Clear figure for next page
        plt.clf()
        time.sleep(120)  # Wait for 2 minutes before showing the next page

        # Check elapsed time and alert if 20 minutes have passed
        elapsed_time = time.time() - start_time
        if elapsed_time >= 1200:  # 1200 seconds = 20 minutes
            print("Alert: 20 minutes have passed since execution started.")
            break  # Optionally stop execution after the alert

    plt.ioff()  # Turn off interactive mode
    plt.close()

if __name__ == "__main__":
    pdf_path = "/content/Ai-Presentation.pdf"
    display_pdf(pdf_path)
```
